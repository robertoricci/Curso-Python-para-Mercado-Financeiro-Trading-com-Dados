#3.Extração dos demonstrativos trimestrais das empresas

Quais demonstrativos você precisa?

Em 'arquivo', determine o tipo de demonstrativo e o ano

Esse 'arquivo' precisa estar contido na pasta zipada da CVM

O 'arquivo' precisa estar na pasta zipada contida em 'link'


```python
import pandas as pd
import requests
from io import BytesIO
from zipfile import ZipFile
```

Função para retornar um dataframe do ano e arquivo desejado !!!
Pois no ZIP tem varios tipos de arquivos !!!

```python

# le um arquivo csv ou txt dentro  de um zip na internet e retorna um dataframe
def read_csv_from_zip(url, file, sep=';'):
    # Requisição do arquivo
    r=requests.get(url)

    # Faz a leitura do arquivo zip
    zf = ZipFile(BytesIO(r.content))

    # Abre o arquivo desejado
    file= zf.open(file)
    
    # Cria uma lista com os dados e decodifica 
    lines = file.readlines()
    #lista das linhas 
    lines=[i.strip().decode('ISO-8859-1') for i in lines]
    #fechar arquivo
    file.close()
    
    # separa por ; e coloca no dataframe
    values = [i.replace('\n','').strip().split(sep) for i in lines]
    #transformar em dataframe
    df = pd.DataFrame(values[1:], columns=values[0])
    return df

  # ano = ano do demonstrativo
  # cod = qual o demonstrativo {DFV,BPA,DMPL,DRA,DRE,DVA}
  # tipo_demonstrativo = 'con' ou 'ind'

def relatorio_cias_abertas(ano, cod, tipo_periodo, tipo_demonstrativo):
    #https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/itr_cia_aberta_2022.zip
    url=f'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/{tipo_periodo}/DADOS/itr_cia_aberta_{ano}.zip'
    arquivo = f'{tipo_periodo.lower()}_cia_aberta_{cod}_{tipo_demonstrativo}_{ano}.csv'
    df = read_csv_from_zip(url, arquivo)
    return df
```

```python
#tipo_periodo ITR -> informação trimestrais
# Balanço Patrimonial Ativo (BPA)
# Balanço Patrimonial Passivo (BPP)
# Demonstração de Fluxo de Caixa - Método Direto (DFC-MD)
# Demonstração de Fluxo de Caixa - Método Indireto (DFC-MI)
# Demonstração das Mutações do Patrimônio Líquido (DMPL)
# Demonstração de Resultado Abrangente (DRA)
# Demonstração de Resultado (DRE)

df = relatorio_cias_abertas(ano=2022,cod='DRE',tipo_periodo='ITR',tipo_demonstrativo='con')
```
