```python
import pandas as pd
import requests
from io import BytesIO
from zipfile import ZipFile
```

Função para retornar um dataframe do ano e arquivo desejado !!!
Pois no ZIP tem varios tipos de arquivos !!!

```python

# le um arquivo csv ou txt dentro  de um zip na internet e retorna um dataframe
def read_csv_from_zip(url, file, sep=';'):
    # Requisição do arquivo
    r=requests.get(url)

    # Faz a leitura do arquivo zip
    zf = ZipFile(BytesIO(r.content))

    # Abre o arquivo desejado
    file= zf.open(file)
    
    # Cria uma lista com os dados e decodifica 
    lines = file.readlines()
    #lista das linhas 
    lines=[i.strip().decode('ISO-8859-1') for i in lines]
    #fechar arquivo
    file.close()
    
    # separa por ; e coloca no dataframe
    values = [i.replace('\n','').strip().split(sep) for i in lines]
    #transformar em dataframe
    df = pd.DataFrame(values[1:], columns=values[0])
    return df

  # ano = ano do demonstrativo
  # cod = qual o demonstrativo {DFV,BPA,DMPL,DRA,DRE,DVA}
  # tipo_demonstrativo = 'con' ou 'ind'

def relatorio_cias_abertas(ano, cod, tipo_periodo, tipo_demonstrativo):
    #https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/itr_cia_aberta_2022.zip
    url=f'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/{tipo_periodo}/DADOS/itr_cia_aberta_{ano}.zip'
    arquivo = f'{tipo_periodo.lower()}_cia_aberta_{cod}_{tipo_demonstrativo}_{ano}.csv'
    df = read_csv_from_zip(url, arquivo)
    return df
```

```python
#tipo_periodo ITR -> informação trimestrais
#tipo_periodo DPC -> informação Anual
# Balanço Patrimonial Ativo (BPA)
# Balanço Patrimonial Passivo (BPP)
# Demonstração de Fluxo de Caixa - Método Direto (DFC-MD)
# Demonstração de Fluxo de Caixa - Método Indireto (DFC-MI)
# Demonstração das Mutações do Patrimônio Líquido (DMPL)
# Demonstração de Resultado Abrangente (DRA)
# Demonstração de Resultado (DRE)
df = relatorio_cias_abertas(ano=2022,cod='DRE',tipo_periodo='ITR',tipo_demonstrativo='con')
```

```python
def dowload_demonstrativo_por_empresa(ano,tipo_periodo,empresas,demonstrativos):
  import time
  start_time = time.time()
  #tipo_periodo(anual..trimestral..) = 'itr'
  #demonstrativos = ['DRE_con']
  #empresas = ['022357']
  lista_listas = []
  #montar url 
  url=f'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/{tipo_periodo}/DADOS/{tipo_periodo.lower()}_cia_aberta_{ano}.zip'
  # Requisição do arquivo
  r=requests.get(url)

  # Faz a leitura do arquivo zip
  zf = ZipFile(BytesIO(r.content))
  a=0
  for em in empresas:
    lista_df = []
    #percorrer as empresas e tipo de demonstrativo

    for de in demonstrativos:
      #arquivo = f'{tipo_periodo.lower()}_cia_aberta_{cod}_{tipo_demonstrativo}_{ano}.csv'
      file=f'{tipo_periodo.lower()}_cia_aberta_{de}_{ano}.csv'
        # Abre o arquivo desejado
      file= zf.open(file)
      # Cria uma lista com os dados e decodifica 
      lines = file.readlines()
        #lista das linhas 
      lines=[i.strip().decode('ISO-8859-1') for i in lines]
        #fechar arquivo
      file.close()
        # faz um split por ; 
      values = [i.replace('\n','').strip().split(';') for i in lines]
        #transformar em dataframe
      df = pd.DataFrame(values[1:], columns=values[0])
      #criar uma coluna nova
      df['VL_AJUSTADO'] = pd.to_numeric(df['VL_CONTA'])

      filtro = df[df['CD_CVM']==str(em).zfill(6)]
      lista_df.append(filtro)

      print(f'Trabalhando com a empresa {em} e seu demonstrativo {de}. As dimensões são {filtro.shape} ')

    lista_listas.append(lista_df)

    # Utilizando a biblioteca xlsxwriter
    writer = pd.ExcelWriter(f'Demonstrativos Empresa {str(em)}.xlsx', engine= 'xlsxwriter')

    for x,dem in enumerate(demonstrativos):
      # Especifique o que esse arquivo de Excel vai conter
      lista_listas[a][x].to_excel(writer, sheet_name = dem, encoding = 'ISO-8859-1',index=False)
    a += 1

    print(f'Arquivo Excel com os demonstrativos da empresa {str(em)} já foi exportado. \n')

    # Fechar o arquivo Excel
    writer.save()

  print("O tempo de execução desse programa foi de %s segundos ---" %(time.time() -start_time))
```

```python

def dowload_empresa_agrupada_por_demonstrativo(ano,tipo_periodo,empresas,demonstrativos):
  import time
  start_time = time.time()
  #preencher com 6 casas o codigo da cvm
  emp=[]
  for em in empresas:
    emp.append(str(em).zfill(6))
  #tipo_periodo = 'itr'
  #demonstrativos = ['DRE_con']
  #empresas = ['022357']
  lista_listas = []
  #montar url 
  url=f'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/{tipo_periodo}/DADOS/{tipo_periodo.lower()}_cia_aberta_{ano}.zip'
  # Requisição do arquivo
  r=requests.get(url)
  # Faz a leitura do arquivo zip
  zf = ZipFile(BytesIO(r.content))
  a=0
    #percorrer as empresas e tipo de demonstrativo
  for de in demonstrativos:
      #arquivo = f'{tipo_periodo.lower()}_cia_aberta_{cod}_{tipo_demonstrativo}_{ano}.csv'
      file=f'{tipo_periodo.lower()}_cia_aberta_{de}_{ano}.csv'
        # Abre o arquivo desejado
      file= zf.open(file)
      # Cria uma lista com os dados e decodifica 
      lines = file.readlines()
        #lista das linhas 
      lines=[i.strip().decode('ISO-8859-1') for i in lines]
        #fechar arquivo
      file.close()
        # faz um split por ; 
      values = [i.replace('\n','').strip().split(';') for i in lines]
        #transformar em dataframe
      df = pd.DataFrame(values[1:], columns=values[0])
      #criar uma coluna nova
      #df.query()
      df['VL_AJUSTADO'] = pd.to_numeric(df['VL_CONTA'])

      df.query("CD_CVM in @emp",inplace=True)

      print(f'Trabalhando o demonstrativo {de}. As dimensões são {df.shape} ')

      writer = pd.ExcelWriter(f'Demonstrativos Empresa {str(de)}.xlsx', engine= 'xlsxwriter')
      df.to_excel(writer)
      writer.save()
      print(f'Arquivo Excel do demonstrativo  {str(de)} já foi exportado. \n')
   
  print("O tempo de execução desse programa foi de %s segundos ---" %(time.time() -start_time))
```

```python
#Baixar todos as empresas por demonstrativos
dowload_empresa_agrupada_por_demonstrativo(2021,'itr',empresas,demonstrativos)
```
