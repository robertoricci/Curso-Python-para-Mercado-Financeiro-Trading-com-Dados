```python
import yfinance as yf
```

```python
ativos = (df_ibovespa['Código'].values+'.SA').tolist()
```

```python
ativos
```

```python
start = '2022-01-01'
end = '2022-11-30'
```

```python
fechamentos = yf.download(tickers = ativos, start = start, end = end, rounding = True)['Adj Close']
```

```python
fechamentos = fechamentos.reset_index()
```

```python
fechamentos['Date'] =  pd.to_datetime(fechamentos['Date'].dt.strftime('%Y-%m-%d'))
```

```python
fechamentos.dtypes
```

```python
fechamentos.set_index('Date',inplace=True)
```

```python
#calcular os retornos
retornos = fechamentos.pct_change()
```

```python
retornos
```

```python
retornos.isna().any(axis=None)
```

```python
retornos.loc[:, retornos.isnull().any()].columns
```

```python
#retirar o SA dos nomes dos ativos e excluir os valores nulos
retornos.columns = retornos.columns.str.rstrip('.SA')
```

```python
#excluir somente os que estão com todas vazias 
retornos.dropna(how='all',inplace=True)
```
