## 4. Realizando um join

```python
from IPython.display import Image
Image(url = 'https://www.thecrazyprogrammer.com/wp-content/uploads/2019/05/Joins-in-SQL-Inner-Outer-Left-and-Right-Join.jpg')
```

```python
lren.head()
```

```python
weg.head()
```

```python
#renomear a colunas
lren.rename(columns = {'Adj Close': 'LREN3'}, inplace = True)
```

```python
lren.head()
```

```python
weg.rename(columns = {'Adj Close': 'WEGE3'}, inplace = True)
```

```python
weg.head()
```

```python
#mergear através do compo date
novo_df = pd.merge(lren, weg, how = 'inner', on = 'Date')
```

```python
#alterar o campo da WEG para datetime, igual ao outro 
weg['Date'] = pd.to_datetime(weg['Date'], format = '%Y-%m-%d')
```

```python
#deixar o campo como index
weg.index = weg['Date']
```

```python
weg.head()
```

```python
weg.drop(['Date'], axis = 1, inplace = True)
```

```python
weg.head()
```

```python
novo_df = pd.merge(lren, weg, how = 'inner', on = 'Date')
```

```python
novo_df.head()
```

```python
#pegar somente as 2 colunas 
cotacoes = novo_df[['LREN3', 'WEGE3']]
```

```python
cotacoes.head()
```

```python
#como foi utilizado o INNER trouxe somente as datas que tem nos 2 ativos!!!
cotacoes.plot();
```
